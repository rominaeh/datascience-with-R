```r
# p 27
# chi square (x,df)
alpha = c(0.005, 0.01, 0.025, 0.05, 0.1, 0.9, 0.95, 0.975, 0.99, 0.995)
df = c(1:20, 30, 40, 50, 60, 70, 80, 90, 100)
ct = matrix(0, nrow=29, ncol=11)
ct[2:29, 1] = df
ct[1, 2:11] = alpha
for (i in 2:29)
for(j in 2:11)
ct[i,j] = qchisq(ct[1,j], df = ct[i,1])
print(rbind(round(ct[1,], 3), round(ct[2:29,], 2)))
```

```r
# 1
# Total number of jobs
total_jobs = 20

# Probability of an applicant being a woman
p_woman = 0.25

# Probability of selecting more than one woman
# We use 1 minus the probability of selecting 0 or 1 woman
prob_more_than_one_woman = 1 - pbinom(1, total_jobs, p_woman)

print(prob_more_than_one_woman)
```

```r
# 2
# Probability of getting sick is 1 - effectiveness
prob_sick = 1 - 0.8

# Number of people vaccinated
n = 5

# Probability of at most one person getting sick
prob_at_most_one_sick = pbinom(1, n, prob_sick)

print(prob_at_most_one_sick)
```

```r
# 21
mean = 50
std = 2
# A - p>51
1-pnorm(51,mean,std)
# ***** B - When taking samples of size 9 from a normal distribution with a mean of 50 kg and a standard deviation of 2 kg, the distribution of the sample mean will also be normal. However, the standard deviation of this sampling distribution (the standard error) will be smaller than the population standard deviation. The standard error is calculated as the population standard deviation ( = 2) divided by the square root of the sample size (n = 9), resulting in /n = 2/9.
pnorm(51.5, mean = 50, sd = 2/sqrt(9)) - pnorm(49, mean = 50, sd = 2/sqrt(9))
# c -
qnorm(0.1,mean,std)
```
