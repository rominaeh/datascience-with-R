missing data handling


```r
pima = read.csv('/content/pima-indians-diabetes.csv')
head(pima)
dim(pima)
summary(pima)
```

```r
pima$Glucose[pima$Glucose == 0] = NA
pima$BloodPressure[pima$BloodPressure == 0] = NA
pima$SkinThickness[pima$SkinThickness == 0] = NA
pima$Insulin[pima$Insulin == 0] = NA
pima$BMI[pima$BMI == 0] = NA
summary(pima)
```

```r
mean(pima$BMI, na.rm=TRUE)
```

```r
pima$Outcome <- factor(pima$Outcome)
levels(pima$Outcome) = c("negative", "positive")
npt = c(sum(pima$Outcome=="positive"), sum(pima$Outcome=="negative"))
npt
```

```r
install.packages("plotrix")
library(plotrix)
pie3D(npt, labels = levels(pima$test),explode = 0.1, main = "Pie Chart ")
```

```r
plot(pima$Outcome, pima$BloodPressure, ylab="blood", main="Boxplot")
```

```r
par(mfrow = c(1,2))
hist(pima$BloodPressure, xlab="Diastolic", main="", col =rainbow(7))
boxplot(pima$BloodPressure, ylab="Diastolic")
par(mfrow = c(1,1))
```

```r
plot(pima$BMI, pima$SkinThickness, xlab="BMI", ylab="Triceps", main="Scatterplot ")
# slight positive correlation
```

```r
pima1 = data.frame(pima$BMI, pima$SkinThickness)
colnames(pima1) = c('BMI', 'SkinThickness')
pima1 = na.omit(pima1)
cor(pima1$BMI, pima1$SkinThickness)
```
